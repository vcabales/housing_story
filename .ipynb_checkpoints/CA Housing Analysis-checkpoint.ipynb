```python
# Take average for 2018

county_rental['2018-avg'] = county_rental[['2018-01','2018-02','2018-03','2018-04','2018-05','2018-06']].mean(axis=1)
county_rental.nlargest(15, '2018-avg')
```

```python
ca_rental['2018-avg'] = ca_rental[['2018-01','2018-02','2018-03','2018-04','2018-05','2018-06']].mean(axis=1)
ca_rental.nlargest(15, '2018-avg')
```

```python
state_rental_years = [] # 2010 to 2017
start = 2011
for i in range(7):
    year = []
    for j in range(12):
        if j < 9:
            month = '0'+str(j+1)
        else:
            month = str(j+1)
        year.append(str(start)+'-'+month)
    start += 1
    state_rental_years.append(year)

state_rental['2010-avg'] = state_rental[['2010-11','2010-12']].mean(axis=1)

curr_year = 2011
for y in state_rental_years:
    state_rental[str(curr_year)+"-avg"] = state_rental[y].mean(axis=1)
    curr_year += 1
state_rental.head()
```

```python
state_rental['2018-avg'] = state_rental[['2018-01','2018-02','2018-03','2018-04','2018-05','2018-06']].mean(axis=1)
top10state_rental = state_rental.nlargest(10, '2018-avg')[['RegionName','2010-avg','2011-avg','2012-avg','2013-avg','2014-avg','2015-avg','2016-avg','2017-avg','2018-avg']]
top10state_rental.to_csv('data/top10_state_rentals.csv')
```

```python
state_sales = pd.read_csv('data/State_Zhvi_SingleFamilyResidence.csv')
state_sales.head()
```

```python
state_sale_years = [] # 2010 to 2017
start = 2010
for i in range(8):
    year = []
    for j in range(12):
        if j < 9:
            month = '0'+str(j+1)
        else:
            month = str(j+1)
        year.append(str(start)+'-'+month)
    start += 1
    state_sale_years.append(year)

curr_year = 2010
for y in state_sale_years:
    state_sales[str(curr_year)+"-avg"] = state_sales[y].mean(axis=1)
    curr_year += 1
state_sales['2018-avg'] = state_sales[['2018-01','2018-02','2018-03','2018-04','2018-05','2018-06']].mean(axis=1)
state_sales.head()
```

```python
state_sales.nlargest(10,'2010-avg')
```

```python
top10state_sales = state_sales.nlargest(10,'2018-avg')[['RegionName','2010-avg','2011-avg','2012-avg','2013-avg','2014-avg','2015-avg','2016-avg','2017-avg','2018-avg']]
top10state_sales.to_csv('data/top10_state_sales.csv')
```
